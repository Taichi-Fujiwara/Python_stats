# 合計値

```python
# 　サンプルデータ群
num = [0,100,20,5,18,92,67,245,44,71]
```

```python
#合計値を求める
data_sum = sum(num)

print("合計値 : ",data_sum)
```

# 平均値

```python
# データ数
n = len(num)

print("データ個数 : ", n)
```

```python
# 平均値
avg = data_sum / n

print("平均値 : ", avg)
```

# 中央値

```python
import statistics

median = statistics.median(num)

print("中央値 : ", median)
```

# 平方和

```python
import math

s = 0

# 平方和
for i in range(n):
  s += math.pow( num[i] - avg ,2)

print("平方和 : ", s)
```

# 分散

```python
# 分散
v = s / (n-1)

print("分散 : ", v)
```

# 標準偏差

```python
# 標準偏差
ss = math.sqrt(v)

print("標準偏差 : ", ss)
```

# 散布図

```python
import matplotlib.pyplot as plt
x = [0,100,20,5,18,92,67,245,44,71]
y = [20,90,30,0,42,2,80,0,55,81]

px = pd.Series(x)
py = pd.Series(y)

rxy =px.corr(py) 
print("相関係数 : ", rxy)


# 散布図を描画
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
```
